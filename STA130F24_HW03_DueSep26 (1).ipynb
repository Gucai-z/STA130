````python
import seaborn as sns
import plotly.graph_objects as go
import numpy as np
penguins = sns.load_dataset('penguins')
penguins = penguins.dropna(subset=['flipper_length_mm'])
species_list = penguins['species'].unique()
fig = go.Figure()
colors = {'Adelie': 'lightblue', 'Gentoo': 'lightgreen', 'Chinstrap': 'lightcoral'}
for species in species_list:
    species_data = penguins[penguins['species'] == species]['flipper_length_mm']
    
  
    mean = species_data.mean()
    median = species_data.median()
    std_dev = species_data.std()
    iqr = np.percentile(species_data, 75) - np.percentile(species_data, 25)
    min_value = species_data.min()
    max_value = species_data.max()


    fig.add_trace(go.Histogram(x=species_data, name=f'{species} Histogram', opacity=0.6, marker_color=colors[species]))

    fig.add_vline(x=mean, line=dict(color='blue', dash='dash'), name=f'{species} Mean', 
                  annotation_text=f"{species} Mean", annotation_position="top left")

    fig.add_vline(x=median, line=dict(color='red', dash='dash'), name=f'{species} Median',
                  annotation_text=f"{species} Median", annotation_position="top right")

    fig.add_vrect(x0=min_value, x1=max_value, fillcolor=colors[species], opacity=0.2, layer="below", line_width=0,
                  annotation_text=f'{species} Range', annotation_position="top left")

    fig.add_vrect(x0=np.percentile(species_data, 25), x1=np.percentile(species_data, 75),
                  fillcolor=colors[species], opacity=0.3, line_width=0, annotation_text=f'{species} IQR')

    fig.add_vrect(x0=mean - 2 * std_dev, x1=mean + 2 * std_dev, fillcolor='yellow', opacity=0.2, layer="below",
                  line_width=0, annotation_text=f'{species} ± 2σ')

fig.update_layout(
    title="Flipper Length Distribution by Species",
    xaxis_title="Flipper Length (mm)",
    yaxis_title="Count",
    legend_title="Legend",
    barmode='overlay'
)

fig.show()
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Create subplots: 1 row and 3 columns
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# List of unique species
species = iris['species'].unique()

# Loop through each species and create a KDE plot
for ax, species_name in zip(axes, species):
    # Filter data for the current species
    species_data = iris[iris['species'] == species_name]
    
    # Create the KDE plot for the current species
    sns.kdeplot(species_data['sepal_length'], ax=ax, bw_adjust=0.5)
    
    # Add title and labels
    ax.set_title(species_name)
    ax.set_xlabel("Sepal Length")
    ax.set_ylabel("Density")

# Adjust layout
plt.tight_layout()
plt.suptitle("KDE Plots of Sepal Length for Different Iris Species", y=1.05)
plt.show()
````

````python
import pandas as pd
import plotly.express as px

# Load the dataset
bn = pd.read_csv('https://raw.githubusercontent.com/hadley/data-baby-names/master/baby-names.csv')

# Make identical boy and girl names distinct
bn['name'] = bn['name'] + " " + bn['sex'] 

# Rank names by year
bn['rank'] = bn.groupby('year')['percent'].rank(ascending=False)
bn = bn.sort_values(['name', 'year'])

# Calculate percent change
bn['percent change'] = bn['percent'].diff()
new_name = [True] + list(bn.name[:-1].values != bn.name[1:].values)
bn.loc[new_name, 'percent change'] = bn.loc[new_name, 'percent']

# Sort values by year
bn = bn.sort_values('year')

# Filter for common names
bn = bn[bn.percent > 0.001] 

# Create the scatter plot
fig = px.scatter(
    bn,
    x='percent change',          # X-axis: Percent change
    y='rank',                    # Y-axis: Rank
    animation_frame='year',      # Animate by year
    animation_group='name',      # Group by name
    size='percent',              # Size of markers: Percent
    color='sex',                 # Color by sex
    hover_name='name',           # Hover text shows name
    size_max=50,                 # Maximum size of markers
    range_x=[-0.005, 0.005]      # X-axis range
)

# Reverse the y-axis
fig.update_yaxes(autorange='reversed')

# Show the plot
fig.show(renderer="png")  # Render for submission
````
